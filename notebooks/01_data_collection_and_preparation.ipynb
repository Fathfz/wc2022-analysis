## Isolate the Final and Load the Event Table

The code then identifies the match shared by France and Argentina in order to isolate the final automatically. This step returns a unique match identifier, here **3869685**, which is then used to download the associated event table with `sb.events()`.

At that point, we obtain a much finer dataset made up of **4407 events**. From there, the work no longer focuses only on scores or match sheets, but on the full granularity of play: passes, controls, carries, duels, pressures, shots, fouls, or recoveries.

</details>


```python
france_matches = matches[
    (matches["home_team"] == "France") | (matches["away_team"] == "France")
].copy()

arg_matches = matches[
    (matches["home_team"] == "Argentina") | (matches["away_team"] == "Argentina")
].copy()

print("Matchs de la France :")
france_matches[["match_date", "home_team", "away_team", "home_score", "away_score"]].sort_values("match_date")
```

```python
print("Matchs de l'Argentine :")
arg_matches[["match_date", "home_team", "away_team", "home_score", "away_score"]].sort_values("match_date")
```

## Comprendre la structure des evenements

Les comptages par type d'evenement servent de controle qualitatif sur les donnees chargees. Les categories les plus frequentes sont les **passes (1263)**, les **ball receipts (1114)** et les **carries (940)**, devant les **pressions (361)** puis les actions plus disruptives comme les duels, dribbles, fautes ou tirs.

Cette hierarchie est coherente avec la nature d'un match de football evenementiel: la circulation du ballon produit naturellement l'essentiel du volume. Le code permet donc ici de verifier que la table de la finale est exploitable et de reperer les variables les plus interessantes pour les notebooks suivants, en particulier pour l'etude des passes, des tirs et des sequences menant au danger offensif.

</details>

## Understanding the Event Structure

The counts by event type serve as a qualitative control on the loaded data. The most frequent categories are **passes (1263)**, **ball receipts (1114)**, and **carries (940)**, ahead of **pressures (361)** and then more disruptive actions such as duels, dribbles, fouls, or shots.

This hierarchy is consistent with the nature of event data in football: ball circulation naturally produces most of the volume. The code therefore helps verify that the final-event table is usable and highlights the most relevant variables for the following notebooks, especially for studying passes, shots, and the sequences leading to attacking danger.

</details>


```python
finale = matches[
    (matches["home_team"].isin(["France", "Argentina"])) &
    (matches["away_team"].isin(["France", "Argentina"]))
].copy()

finale[["match_id", "match_date", "home_team", "away_team", "home_score", "away_score"]]
```

```python
finale_id = finale["match_id"].values[0]
print(f"ID de la finale : {finale_id}")
```

```python
events_finale = sb.events(match_id=finale_id)

print(f"Nombre d'événements : {len(events_finale)}")
events_finale.head()
```
